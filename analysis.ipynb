```python
data = pd.read_sql(joined_query, con=engine)
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data = data.drop_duplicates
```

```python
purchase_data_query = '''
select * from transaction_data;
'''
transaction_data = pd.read_sql(purchase_data_query, con=engine)
```

```python
merged_df = transaction_data.merge(purchase_data, on='LYLTY_CARD_NBR', how='left')
```

```python
merged_df.head()
```

### objectives
1. top 10 most purchased brand
2. total sales by customer life stage
3. average spend per customer category

### Top 10 brands

```python
top_brands = (merged_df.groupby('PROD_NAME')['PROD_QTY'].sum().sort_values(ascending=False).head(10))
```

```python
top_brands
```

### total sales by customer life stage

```python
sales_by_lifestage = (merged_df.groupby('LIFESTAGE')['TOT_SALES'].sum().sort_values(ascending=False))
```

```python
sales_by_lifestage
```

### average spend per customer category

```python
avg_spend_category = (merged_df.groupby('PREMIUM_CUSTOMER')['TOT_SALES'].mean())
```

```python
avg_spend_category
```

### Monthly sales trend

```python
merged_df['YEAR']=merged_df['DATE'].dt.year
```

```python
merged_df['MONTH']=merged_df['DATE'].dt.month
```

```python
merged_df['MONTH_NAME']=merged_df['DATE'].dt.month_name()
```

```python
merged_df.head()
```

```python
monthly_sales=(merged_df.groupby(['YEAR', 'MONTH_NAME'])['TOT_SALES'].sum().reset_index().sort_values(['YEAR', 'MONTH_NAME']))
```

```python
monthly_sales
```

```python
yearly_sales=(merged_df.groupby('YEAR')['TOT_SALES'].sum().sort_values(ascending=False))
```

```python
yearly_sales
```

### store performance

```python
store_sales = (merged_df.groupby('STORE_NBR')['TOT_SALES'].sum().sort_values(ascending=False).head(10))
```

```python
store_sales
```

```python
product_sales = (merged_df.groupby('PROD_NAME')['TOT_SALES'].sum().sort_values(ascending=False).head(10))
```

```python
product_sales
```
